```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\theka\OneDrive\Desktop\householdtask3.csv'
df = pd.read_csv(file_path)

print("Initial rows of the DataFrame:")
print(df.head())
```

```python
mean_values = df.mean(numeric_only=True)

plt.figure(figsize=(10, 6))
mean_values.plot(kind='bar', color='skyblue')
plt.title('Mean Values of Numerical Columns')
plt.xlabel('Columns')
plt.ylabel('Mean Value')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
if len(df.columns) >= 2:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df.iloc[:, 1], label=df.columns[1], color='blue', marker='o')
    plt.plot(df.index, df.iloc[:, 2], label=df.columns[2], color='red', marker='x')
    plt.title('Line Chart of First Two Numerical Columns')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("The DataFrame does not have enough numerical columns for the line chart.")
```
