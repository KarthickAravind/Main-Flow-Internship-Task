```python
import pandas as pd

file_path = r'01.Data Cleaning and Preprocessing.csv'
df = pd.read_csv(file_path)
print(df.head())
```

```python
# Filter data based on conditions
filtered_df = df[df['BlowFlow'] > 1300]
print("\nFiltered DataFrame where 'BlowFlow' > 1300:")
print(filtered_df)
```

```python
# Handle missing values by filling with the mean
df_filled = df.fillna(df.mean(numeric_only=True))
print("\nDataFrame after filling missing values with mean:")
print(df_filled)
```

```python
# Calculate summary statistics
summary_stats = df.describe()
print("\nSummary statistics for the numerical columns:")
print(summary_stats)
```
